#**Transformando dados em uma tabela**

##Observando nas tabelas mostradas, vemos que esta em metodo json, e para limpar e transformar esses dados, irei utilizar o metodo normalize junto com a biblioteca json

```python
with open('dataset-telecon.json') as f:
  json_bruto = json.load(f)
  json_bruto
```

```python
df = pd.json_normalize(json_bruto)
df.head()
```

```python
df.info()
```

```python
df['conta.cobranca.Total'] = df['conta.cobranca.Total'].astype(float)
```

```python
df[df['conta.cobranca.Total'] == ' '].head()
```

```python
df[df['conta.cobranca.Total'] == ' '][
    ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']
]
```

```python
idx = df[df['conta.cobranca.Total'] == ' '].index
```

```python
df.loc[idx, 'conta.cobranca.Total'] = df.loc[idx, 'conta.cobranca.mensal'] * 24
```

```python
df.loc[idx, 'cliente.tempo_servico'] = 24
```

```python
df.loc[idx][
    ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']
]
```

```python
df['conta.cobranca.Total'] = df['conta.cobranca.Total'].astype(float)
```

#**Trabalhando com os Outliers**

```python
df_sem_vazio.describe()
```

```python
import seaborn as sns
```

```python
#plotar para ver se tem outliers
sns.boxplot(x=df_sem_vazio['cliente.tempo_servico']);
```

```python
Q1 = df_sem_vazio['cliente.tempo_servico'].quantile(.25)
Q3 = df_sem_vazio['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1-1.5*IQR
limite_superior = Q3 +1.5*IQR
```

```python
outliers_index = (df_sem_vazio['cliente.tempo_servico'] < limite_inferior) | (df_sem_vazio['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
df_sem_vazio[outliers_index]['cliente.tempo_servico']
```

#**Substituindo valores para os outliers**

```python
#fazer a copia para um novo dataframe
df_sem_out = df_sem_vazio.copy()
```

```python
df_sem_out[outliers_index]['cliente.tempo_servico']
```

```python
df_sem_out.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
    df_sem_out.loc[outliers_index, 'conta.cobranca.Total'] / df_sem_out.loc[outliers_index, 'conta.cobranca.mensal']
)
```

```python
sns.boxplot(x=df_sem_out['cliente.tempo_servico']);
```

```python
df_sem_out[outliers_index][['cliente.tempo_servico','conta.cobranca.mensal','conta.cobranca.Total']]
```

#**Removendo Outliers**

##Remover os outliers para o machine learning ser mais preciso

```python
df_sem_out[outliers_index]['cliente.tempo_servico']
```

```python
Q1 = df_sem_out['cliente.tempo_servico'].quantile(.25)
Q3 = df_sem_out['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1-1.5*IQR
limite_superior = Q3 +1.5*IQR

outliers_index = (df_sem_out['cliente.tempo_servico'] < limite_inferior) | (df_sem_out['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
df_sem_out[outliers_index]
```

```python
df_sem_out = df_sem_out[~outliers_index]
df_sem_out
```

```python
df_sem_out.reset_index(drop=True, inplace=True)
```

```python
sns.boxplot(x=df_sem_out['cliente.tempo_servico']);
```
